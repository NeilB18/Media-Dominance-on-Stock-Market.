```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    sentiment_score = sia.polarity_scores(text)['compound']
    return round((sentiment_score + 1) / 2, 2)  # Map the sentiment score to a range of 0.0 to 1.0

def add_sentiment_scores(dataset):
    dataset['SentimentScore'] = dataset['Headline'].apply(analyze_sentiment)
    return dataset

if __name__ == "__main__":
    # Replace 'your_dataset.csv' with the actual path to your dataset file.
    dataset = pd.read_excel('S&P500_News.xlsx')
    dataset_with_sentiment = add_sentiment_scores(dataset)
    dataset_with_sentiment.to_excel('dataset_with_sentiment.xlsx', index=False)
    print("Sentiment analysis completed and dataset updated!")
```
