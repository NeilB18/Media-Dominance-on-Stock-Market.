```python
import pandas as pd
df = pd.read_excel('/Users/neil/Documents/code/Research-Project/data/stock_data/S&P500_data.xlsx')
```

```python
df
```

```python
x = df.loc[:,['Avg. Sentiment Score','Close*']].values
x
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
wcss = []
for i in range(1,21):
    kmeans = KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,21),wcss)
plt.title('The elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2,init="k-means++",random_state=42)
y_pred = model.fit_predict(x)
```

```python
plt.scatter(x[y_pred ==0 , 0],x[y_pred ==0 , 1],s = 100, color = "red", label="Cluster 0")
plt.scatter(x[y_pred ==1 , 0],x[y_pred ==1 , 1],s = 100, color = "blue", label="Cluster 1")
plt.title('Cluster of Avg. Sentiment Score to Closing Price')
plt.xlabel('Avg. Sentiment Score')
plt.ylabel('Close* Price')    
plt.legend()
plt.show()
```
